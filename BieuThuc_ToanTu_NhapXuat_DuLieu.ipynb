```python
a = int(input("Nhập số nguyên a: ")) #a=5
b = int(input("Nhập số nguyên b: ")) #b=2
print("Tổng:", a + b)
print("Hiệu:", a - b)
print("Tích:", a * b)
print("Phần nguyên:", a // b)
print("Phần dư:", a % b)
print("Kết quả chia thực:", format(a / b, ".2f")) # có thể dùng hàm round(a / b, 2)
```
