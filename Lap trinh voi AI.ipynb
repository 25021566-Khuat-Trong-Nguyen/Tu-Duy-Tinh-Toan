```python
#Đoạn code tính tổng các số từ 1 đến 4,Kết quả cuối cùng bằng 10.
n = 4 #Gán giá trị 4 cho biến n.
t = 0 #Khởi tạo biến t để lưu tổng.
for i in range(1, n+1): #Vòng lặp từ 1 đến n (bao gồm n).
    t += i #Cộng giá trị i vào biến t.
print(t) #Kết quả cuối cùng bằng 10.
```

```python
#Chương trình kiểm tra từng số trong danh sách [1,2,3,4] để xác định là chẵn hay lẻ.
numbers = [1, 2, 3, 4] #Danh sách các số cần kiểm tra.
for n in numbers: #Vòng lặp qua từng số trong danh sách.
    if n % 2 == 0: #Kiểm tra nếu số chia hết cho 2.
        print(n, "là số chẵn.") #In ra nếu số là chẵn.
    else: #Nếu số không chia hết cho 2.
        print(n, "là số lẻ.") #In ra nếu số là lẻ. 
```

```python
#Chương trình đếm số phần tử trong danh sách animals,Kết quả cuối cùng là 4, đúng với số lượng phần tử có trong danh sách.
animals = ["cat", "dog", "cat", "bird"] #Danh sách các con vật.
count = 0 #Khởi tạo biến count để đếm số phần tử.
for a in animals: #Vòng lặp qua từng phần tử trong danh sách.
    count += 1 #Tăng biến count lên 1 cho mỗi phần tử.
print(count) #In ra kết quả cuối cùng.
```

    === AI Prediction System ===  
    1) Sentiment analysis  
    2) Weather forecast  
    3) Exit  
    Please choose an option:  
Đây không phải là code Python hoàn chỉnh, mà là nội dung hiển thị của một chương trình mô phỏng hệ thống dự đoán AI (AI Prediction System).
Chương trình này có thể là phần giao diện (menu) để người dùng lựa chọn một chức năng AI nào đó.
Ở dưới là chương trình menu - AI Prediction System – chạy được.

```python
# === AI Prediction System ===
# Chương trình mô phỏng menu chọn chức năng AI
print("=== AI Prediction System ===")
print("1) Sentiment analysis")
print("2) Weather forecast")
print("3) Exit")
# Nhập lựa chọn từ người dùng
choice = input("Please choose an option: ")
# Kiểm tra lựa chọn
if choice == "1":
    print("Bạn đã chọn: Phân tích cảm xúc (Sentiment Analysis)")
    print("AI đang phân tích cảm xúc từ văn bản...")
elif choice == "2":
    print("Bạn đã chọn: Dự báo thời tiết (Weather Forecast)")
    print("AI đang dự đoán thời tiết cho khu vực bạn chọn...")
elif choice == "3":
    print("Thoát chương trình. Tạm biệt!")
else:
    print("Lựa chọn không hợp lệ. Vui lòng chọn 1, 2 hoặc 3.")
```
