```python
#Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu
a = float(input("Nhập vào chiều rộng a của khu đất hình chữ nhật: ")) #a=20
b = float(input("Nhập vào chiều dài b của khu đất hình chữ nhật: ")) #b=30
pi = 3.14
dien_tich_khu_dat = a * b
ban_kinh_hinh_tron = min(a, b) / 2
dien_tich_khu_vui_choi = pi * ban_kinh_hinh_tron ** 2
dien_tich_trong_cay = dien_tich_khu_dat - dien_tich_khu_vui_choi
print("Diện tích còn để trồng cây xung quanh là:", round(dien_tich_trong_cay, 2))
```

```python
#Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.
a = float(input("Nhập vào cạnh a: ")) #10
b = float(input("Nhập vào cạnh b: ")) #20
c = float(input("Nhập vào cạnh c: ")) #25
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    dien_tich = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print("Diện tích của tam giác là:", round(dien_tich, 1))
else:
    print("Khong phai 3 canh cua tam giac")
```

```python
ten_chu_ho = input("Nhập vào họ tên chủ hộ: ") #NGUYEN XUAN HAI
chi_so_thang_truoc = int(input("Nhập vào chỉ số điện kế tháng trước: ")) #2000
chi_so_thang_nay = int(input("Nhập vào chỉ số điện kế tháng này: ")) #2586
so_kwh_tieu_thu = chi_so_thang_nay - chi_so_thang_truoc
tien_dien = 0
if so_kwh_tieu_thu <= 50:
    tien_dien = so_kwh_tieu_thu * 1984
elif so_kwh_tieu_thu <= 100:
    tien_dien = (50 * 1984) + (so_kwh_tieu_thu - 50) * 2050
elif so_kwh_tieu_thu <= 200:
    tien_dien = (50 * 1984) + (50 * 2050) + (so_kwh_tieu_thu - 100) * 2380
elif so_kwh_tieu_thu <= 300:
    tien_dien = (50 * 1984) + (50 * 2050) + (100 * 2380) + (so_kwh_tieu_thu - 200) * 2998
elif so_kwh_tieu_thu <= 400:
    tien_dien = (50 * 1984) + (50 * 2050) + (100 * 2380) + (100 * 2998) + (so_kwh_tieu_thu - 300) * 3350
else:
    tien_dien = (50 * 1984) + (50 * 2050) + (100 * 2380) + (100 * 2998) + (100 * 3350) + (so_kwh_tieu_thu - 400) * 3460
tien_dien = tien_dien + tien_dien * 0.08
print("Họ và tên:", ten_chu_ho)
print("Tiền phải trả là:", round(tien_dien))
```
